```python
# import packages
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
```

```python
# set the device
device = torch.device("mps" if torch.mps.is_available() else "cpu")
print(f"Training with device: {device}")
```

```python
# get the dataset

# set path variable
data = "./data"

# define transforms
train_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# get data
train_dataset = torchvision.datasets.CIFAR10(
    root = data,
    train = True,
    download = True,
    transform = train_transform
)

val_dataset = torchvision.datasets.CIFAR10(
    root = data,
    train = False,
    download = True,
    transform = val_transform
)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')


# get loaders
train_loader = DataLoader(train_dataset, batch_size = 128, shuffle = True)
val_loader = DataLoader(val_dataset, batch_size = 128, shuffle = False)
```

```python
# load model, optimizer, and loss function

# model
from models.resnet import ResidualNN
model = ResidualNN()

# optimizer
optimizer = optim.Adam(model.parameters(),lr = 0.001, weight_decay = 1e-4)

# loss
loss_function = nn.CrossEntropyLoss()
```

```python
# define the training loop
from pathlib import Path
from utilities.train import train_step, evaluate

num_epochs = 20
best_model = None
best_accuracy = 0

for epoch in range(num_epochs) :

    # instantiate training progress bar
    pbar = tqdm(train_loader, desc = f"Epoch {epoch + 1}/{num_epochs}", leave = True)

    # train 1 epoch on training set
    model, avg_epoch_loss, train_accuracy = train_step(model, loss_function, optimizer,device, pbar)

    # evaluate on val test
    val_accuracy = evaluate(model, val_loader, device)
    print(f"Training Accuracy: {train_accuracy * 100 :.2f} % Validation Accuracy: {val_accuracy * 100:.2f} %")

    if val_accuracy > best_accuracy :
        best_model = model
        best_acccuracy = val_accuracy

# save the best model
save_dir= Path("model_params")
best_model_path = save_dir / "my_best_model.pt"
torch.save(best_model.state_dict(), best_model_path)
```
